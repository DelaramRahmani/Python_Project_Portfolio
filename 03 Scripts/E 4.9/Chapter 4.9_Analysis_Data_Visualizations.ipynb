# Load Your Dataset

```python
import pandas as pd
import os

# Define the path
path = '/Users/dela/Documents/15-01-2025 Instacart Basket Analysis/02 Data/Prepared Data'

# Load the dataset
ords_prods_merge = pd.read_pickle(os.path.join(path, 'ords_prods_merge_final.pkl'))

# Display the first few rows
ords_prods_merge.head()
```

# Creating Visualizations

```python
print(ords_prods_merge.columns)
```

# Bar Chart

```python
bar_chart = ords_prods_merge['order_day_of_week'].value_counts().plot.bar(color='skyblue')
plt.title('Orders by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Orders')
plt.show()
```

# Histogram for prices

```python
# Create a histogram for the 'prices' column
ords_prods_merge['prices'].plot.hist(bins=25, color='green')
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

# Scatterplot for Outliers in prices

```python
# Scatterplot for 'prices' column
sns.scatterplot(x='prices', y='prices', data=ords_prods_merge, color='red')
plt.title('Scatterplot of Prices')
plt.xlabel('Price')
plt.ylabel('Price')
plt.show()
```

# Line Chart for Trends

```python
# Line chart for 'prices' across 'order_day_of_week'
line_chart = sns.lineplot(data=ords_prods_merge, x='order_day_of_week', y='prices', ci=None)
plt.title('Average Price by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Price')
plt.show()
```
