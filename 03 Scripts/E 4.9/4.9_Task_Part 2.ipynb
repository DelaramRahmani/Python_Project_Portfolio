# Step 1: Load the cleaned customer data which was already merged with the Instacart dataset

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set the path to your prepared data folder
path = '/Users/dela/Documents/15-01-2025 Instacart Basket Analysis/02 Data/Prepared Data'

# Load the most up-to-date project data
combined_data = pd.read_pickle(os.path.join(path, 'instacart_customer_combined.pkl'))

# Preview the dataset to confirm successful loading
print(combined_data.head())
```

# Step 2: Visualizations

## Bar Chart: Distribution of Orders by Loyalty

```python
# Bar chart for the "loyalty_flag" column
loyalty_bar = combined_data['loyalty_flag'].value_counts().plot.bar(color='skyblue')
plt.title('Distribution of Orders by Loyalty')
plt.xlabel('Loyalty Category')
plt.ylabel('Count of Orders')
plt.show()
```

## Histogram: Distribution of Orders by Hour of Day

```python
# Histogram for "order_hour_of_day"
plt.figure(figsize=(10, 6))
sns.histplot(combined_data['order_hour_of_day'], bins=24, kde=False, color='purple')
plt.title('Distribution of Orders by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Frequency')
plt.show()
```

## Line Chart: Prices by Hour of Day

```python
# Take a random 5% sample of the data
sample_data = combined_data.sample(frac=0.05, random_state=42)

# Line chart for "prices" by "order_hour_of_day"
line_plot = sns.lineplot(data=sample_data, x='order_hour_of_day', y='prices', ci=None)
plt.title('Prices by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Prices')
plt.show()
```

## Scatterplot: Age vs Income

```python
# Scatterplot for "Age" vs "income"
scatter_plot = sns.scatterplot(data=combined_data, x='Age', y='income', alpha=0.6, color='green')
plt.title('Age vs Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

# Step 3: Histogram of order_hour_of_day

```python
# Create a histogram for 'order_hour_of_day'
plt.figure(figsize=(10, 6))
sns.histplot(combined_data['order_hour_of_day'], bins=24, kde=False, color='skyblue')
plt.title('Distribution of Orders by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Frequency')
plt.show()
```

This histogram visualizes the distribution of orders across the 24 hours of the day. Peaks in certain hours indicate the most active times when customers place orders. This insight can guide staffing and delivery scheduling.

# Step 4: Bar Chart of loyalty_flag

```python
# Bar chart for 'loyalty_flag'
loyalty_bar = combined_data['loyalty_flag'].value_counts().plot.bar(color='purple')
plt.title('Distribution of Orders by Loyalty Category')
plt.xlabel('Loyalty Flag')
plt.ylabel('Number of Orders')
plt.show()
```

This bar chart displays the distribution of orders among different loyalty categories, such as "Loyal Customers" and "New Customers." It highlights the relative engagement of customers across categories.

# Step 5: Line Chart - prices by order_hour_of_day


```python
# Sampling 5% of the data for efficiency
sample_data = combined_data.sample(frac=0.05, random_state=42)

# Create a line chart for 'prices' by 'order_hour_of_day'
plt.figure(figsize=(10, 6))
sns.lineplot(data=sample_data, x='order_hour_of_day', y='prices', ci=None, color='green')
plt.title('Average Prices by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Price')
plt.show()
```

This line chart demonstrates the fluctuation in average prices of orders throughout the day. Observing any significant peaks or troughs may provide insights into customer spending behavior at different times.

# Step 6: Line Chart - age vs. number of dependents


```python
# Line chart for 'age' and 'n_dependants'
plt.figure(figsize=(10, 6))
sns.lineplot(data=combined_data, x='Age', y='n_dependants', ci=None, color='orange')
plt.title('Age vs. Number of Dependents')
plt.xlabel('Age')
plt.ylabel('Number of Dependents')
plt.show()
```

This line chart reveals the relationship between customer age and the number of dependents. Observing trends here could help marketers better target customers based on family size.

# Step 7: Scatterplot - age vs. income

```python
# Scatterplot for 'age' vs. 'income'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=combined_data, x='Age', y='income', alpha=0.6, color='blue')
plt.title('Age vs. Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

This scatterplot illustrates the relationship between age and income. Any visible patterns, such as a positive correlation, may indicate how income levels change with age.
